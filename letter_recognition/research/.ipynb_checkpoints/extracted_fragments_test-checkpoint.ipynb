```python
def load_model(filename):
    model = ConvDenseNet(12, 20, 0.5, 37, True).cuda()
    model.load_state_dict(torch.load(filename))
    return model

def encode_class(class_num):
    char_list = [' '] + list(string.ascii_uppercase + string.digits)
    return char_list[class_num]

def plot_output(image_data, output_data):    
    output_ind = np.argsort(output_data)
    for i in range(image_data.shape[0]):
        plt.imshow(image_data[i, :, :, :].reshape(32, 32), cmap='gray', vmin=0, vmax=255)               
        ind1, ind2, ind3 = output_ind[i, -1], output_ind[i, -2], output_ind[i, -3]
        print(encode_class(ind1), encode_class(ind2), encode_class(ind3))
        prob_func = (lambda x: np.exp(output_data[i, x]))
        print(prob_func(ind1), prob_func(ind2), prob_func(ind3))
        plt.show()
        
def plot_letter(image_data, output_data, letter, thres):
    output_ind = np.argsort(output_data)
    for i in range(image_data.shape[0]):
        ind = output_ind[i, -1]
        if encode_class(ind) != letter:
            continue
        if np.exp(output_data[i, ind]) < thres:
            continue
        plt.imshow(image_data[i, :, :, :].reshape(32, 32), cmap='gray', vmin=0, vmax=255)               
        plt.show()
    
def plot_lettters(image_data, output_data, thres):
    for letter in list(string.ascii_uppercase + string.digits):
        print(letter)
        plot_letter(image_data, output_data, letter, thres)
        
def test_model(model, filename, from_range, to_range):
    fragment_list = pickle.load(open(filename, 'rb'))
    image_list = map(lambda x: x[1], fragment_list)
    image_data = np.array(image_list, dtype=np.float64).reshape(-1, 1, 32, 32)
    image_data = image_data[from_range:to_range, :, : :]
    var_data = Variable(torch.FloatTensor(image_data).cuda())
    output_data = model(var_data)
    output_data = output_data.cpu().data.numpy()
#    plot_output(image_data, output_data)
    plot_lettters(image_data, output_data, 0.99)

model = load_model('state_dicts/in_place4.pt')
model.eval()
test_model(model, 'key_images/key0.p', 0, 1000)
```

```python
rotate_generator = ImageGenerator(150, False, True)
trained_list = []
for i in range(100):
    trained_list.append(rotate_generator.generate_image()[1])
image_data = np.array(trained_list, dtype=np.float64).reshape(-1, 1, 32, 32)
var_data = Variable(torch.FloatTensor(image_data).cuda())
output_data = model(var_data)
output_data = output_data.cpu().data.numpy()
plot_output(image_data, output_data)
```
